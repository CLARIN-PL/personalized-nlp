```python
import os

import pandas as pd
```

```python
data_dir = os.path.join('..', 'storage', 'data', 'mhs', 'data_raw')

dev_df = pd.read_csv(os.path.join(data_dir, 'dev.tsv'), sep='\t')
test_df = pd.read_csv(os.path.join(data_dir, 'test.tsv'), sep='\t')
train_df = pd.read_csv(os.path.join(data_dir, 'train.tsv'), sep='\t')

dev_df['split'] = 'val'
test_df['split'] = 'test'
train_df['split'] = 'train'
```

```python
for ind, df in enumerate([dev_df, test_df, train_df]):

    if ind == 0:
        full_df = df

    else:
        full_df = pd.concat([full_df, df])

full_df = full_df.rename(columns = {'username': 'annotator'})
```

```python
def set_id(df, column_name) -> pd.DataFrame:
    values_list = df[column_name].unique().tolist()

    df[f'{column_name}_id'] = df[column_name].apply(lambda x: values_list.index(x))
    
    return df
```

```python
full_df = set_id(full_df, 'text')

full_df = set_id(full_df, 'annotator')
```

```python
for dimension in ['sentiment', 'insult', 'humiliate', 'violence', 'hatespeech']:
    for dataset_type in ['train', 'val', 'test']:
        print(f'value count for {dimension} {dataset_type}:', 
              f'{full_df[full_df["split"] == dataset_type][dimension].value_counts()}', 
              sep='\n')
```

```python
save_dir = os.path.join('..', 'storage', 'data', 'mhs')

data_columns = ['text_id', 'text']
annotations_columns = ['text_id', 'annotator_id', 'sentiment', 'respect', 'insult', 'humiliate', 
                       'dehumanize', 'violence', 'genocide', 'hatespeech','split']

full_df[data_columns].to_csv(os.path.join(save_dir, 'data.csv'), index=False)
full_df[annotations_columns].to_csv(os.path.join(save_dir, 'annotations.csv'), index=False)
```
